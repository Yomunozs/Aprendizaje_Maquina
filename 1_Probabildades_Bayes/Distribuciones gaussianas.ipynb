$ P(X|Y=1) = N(X|\mu_x = \mu_1, \sigma_x^2 = \sigma_1^2) \\
  P(X|Y=2) = N(X|\mu_x = \mu_2, \sigma_x^2 = \sigma_2^2) \\
  N_1 = 100 \\
  N_2 = 100 \\
  \mu_1 = 0.5 \\
  \mu_2 = 1 \\
  \sigma_1 = 0.8 \\
  \sigma_2 = 0.6 $

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Valores de X
N_1 = 100
N_2 = 100
mu_1x = 0.5
mu_2x = 2
var_1x = 0.8
var_2x = 0.6

# Valores de Y
mu_1y = 0.5
mu_2y = 1.5
var_1y = 0.15
var_2y = 0.15
```

```python
X_1 = np.random.normal(mu_1x, var_1x, N_1)
X_2 = np.random.normal(mu_2x, var_2x, N_2)
Y_1 = np.random.normal(mu_1y, var_1y, N_1)
Y_2 = np.random.normal(mu_2y, var_2y, N_2)
```

```python
plt.scatter(X_1, Y_1)
plt.scatter(X_2, Y_2)
```

```python
def proba_param (X1, X2, Y1, Y2):
  plt.subplot(2, 2, 1)
  plt.scatter (X1, Y1)
  plt.scatter (X2, Y2)

  plt.subplot(2, 2, 2)
  plt.hist ([np.concatenate((Y1,Y2))], bins = 2, alpha = 0.5)

  plt.subplot(2, 2, 3)
  plt.hist ([np.concatenate((X1, X2))], bins = 20, alpha = 0.5)
  plt.xlim(-2 , 5)

  plt.subplot(2, 2, 4)
  plt.hist (X2, bins = 9, alpha = 0.5)
  plt.xlim(-2, 5)

  plt.show()
```

```python
proba_param(X_1, X_2, Y_1, Y_2)
```
